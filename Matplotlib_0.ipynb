## Line Plot

This type of plot is often used to visualize a trend in data over intervals of time - a **time series**

```python
import matplotlib.pyplot as plt

years = [1983, 1984, 1985, 1986, 1987]
total_populations = [8939007, 8954518, 8960387, 8956741, 8943721]

plt.plot(years, total_populations)
plt.title("Year vs Population in Bulgaria")
plt.xlabel("Year")
plt.ylabel("Total Population")
plt.show()
```

## Scatter plot

* Shows all individual data points which arenâ€™t connected with lines
* This type of plot can be used to display trends or correlations


```python
import matplotlib.pyplot as plt

temp = [30, 32, 33, 28.5, 35, 29, 29]
ice_creams_count = [100, 115, 115, 75, 125, 79, 89]

plt.scatter(temp, ice_creams_count)
plt.title("Temperature vs. Sold ice creams")
plt.xlabel("Temperature")
plt.ylabel("Sold ice creams count")
plt.show()
```

## Histogram

* An accurate representation of the distribution of numeric data
1. Divide the entire range of values into a series of intervals
2. Count how many values fall into each interval (bin)

```python
import matplotlib.pyplot as plt

numbers = [0.1, 0.5, 1, 1.5, 2, 4, 5.5, 6, 8, 9]

plt.hist(numbers, bins = 3)
plt.xlabel("Number")
plt.ylabel("Frequency")
plt.show()
```

*We can see from the histogram above that there are:*
- *5 values(frequency) between 0 and 3*
- *3 values(frequency) between 3 and 6 (including)*
- *2 values(frequency) between 6 (excluding) and 9*

## Bar chart
* Represents categorical data with rectangular bars
* Useful when we want to compare a given numeric value on different categories


```python
languages = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
performance = [10,8,6,4,2,1]
plt.bar(languages, performance)
plt.ylabel('Rating')
plt.title('Programming language usage')
plt.show()
```

## Box plot

A way to show the distribution of values based on the five-number summary:
* Minimum
* First quartile
  - For example, if we have the numbers `1, 3, 4, 7, 8, 8, 9` the first quartile is the median from the `1, 3, 4` values, so itâ€™s `3`
* Median
  - For example, if we have the numbers `1, 2, 5, 6, 8, 9` your median will be `(5 + 6) / 2 = 5`, so it's `5`
* Third quartile
  - For example, if we use these numbers `1, 3, 4, 7, 8, 8, 9` again, the third quartile is the median from the `8, 8, 9` values, so itâ€™s `8`

* IQR (Interquartile Range) = Third quartile - First quartile
* Maximum

boxplt.png

* Outliers:
  - Lies outside the overall pattern
  - Visualized as circles
  - $Third quartile + 1.5 * IQR < Outlier < First quartile - 1.5 * IQR$ 

```python
values = [1, 2, 5, 6, 6, 7, 7, 8, 8, 8, 9, 10, 21]

plt.boxplot(values)
plt.yticks(range(1, 22))
plt.title("Detect outliers for the given values")
plt.ylabel("Value")
plt.show()
```

*In this example the values `1, 2, and 21` are outliers, the maximum is `10`, the minimum is `5`. And, the mean is `7`.*

## Pie chart

* A circular plot
* Divided into slices to show numerical proportion
* Widely used in the business world



```python
sizes = [25, 20, 45, 10]
labels = ["Cats", "Dogs", "Tigers", "Goats"]

plt.pie(sizes, labels = labels, autopct = "%.2f")
plt.axis('equal')
plt.show()
```
