```python
import matplotlib.pyplot as plt

minutes = [1,2,3,4,5,6,7,8,9]

Ronnie = [0,1,1,1,2,3,2,4,4]
Virat = [0,1,2,2,2,4,4,4,4]
Ronaldo = [8,6,5,5,4,2,1,1,0]

labels = ['Ronnie', 'Virat','Ronaldo']
colors = ['green','red','blue']
plt.stackplot(minutes, Ronnie, Virat, Ronaldo, labels = labels, colors = colors)
plt.legend()
plt.title("Stack Plot")
plt.show()
```

```python
plt.style.use("fivethirtyeight")
plt.stackplot(minutes, Ronnie, Virat, Ronaldo, labels = labels, colors = colors)
plt.legend(loc = (0.1,0.1))
plt.title("Stack Plot")
plt.tight_layout()
plt.show()
```
