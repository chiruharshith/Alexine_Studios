```python
import matplotlib.pyplot as plt

minutes = [1,2,3,4,5,6,7,8,9]

Ronnie = [0,1,1,1,2,3,2,4,4]
Virat = [0,1,2,2,2,4,4,4,4]
Ronaldo = [8,6,5,5,4,2,1,1,0]

labels = ['Ronnie', 'Virat','Ronaldo']
```

**ITERATE THROUGH BOTH LISTS AT A TIME**

```python
monday = ['red ball','green ball','blue ball']
tuesday = ['red ball','green ball','yellow ball']

for each in monday:
  print(each)

print("++++++++++++++++")
```

```python
for each in tuesday:
  print(each)
```

```python
for each in zip(monday, tuesday):
  print(each)
```

```python
for each in zip(Ronnie,Virat,Ronaldo):
  plt.pie(each, labels = labels)
  plt.legend()
  plt.title(each)
  plt.show()
```
