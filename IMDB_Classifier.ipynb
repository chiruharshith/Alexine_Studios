### Importing Required Packages

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np
```

```python
# Load The Data
df = pd.read_csv('datasets/imdb_movie.csv')
df.head()
```

**Data Pre-Processing**

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df = df.drop(columns=['director_name','movie_title', 'Genre','facenumber_in_poster', 'movie_imdb_link', 'language', 'title_year', 'aspect_ratio','gross','budget','director_facebook_likes', 'actor_3_facebook_likes','actor_1_facebook_likes'])
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
targets = le.fit_transform(df['imdb_score'])
targets.shape
```

```python
features = df.iloc[:,:-1]
features
```

**Split the data into train and test sets**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

**Classification**

```python
from sklearn.metrics import accuracy_score
dt = DecisionTreeClassifier(max_depth=5)
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
accuracy_score(y_test, y_pred)
```
